```python
import numpy as np
import cv2
from matplotlib import pyplot as plt
%matplotlib inline

# all impls are here:
# see ops.py for impl details
from ops import *

img_name = "lenna.png"
img = cv2.imread("lenna.png", cv2.IMREAD_GRAYSCALE)
print(img.shape, img.dtype)
plt.imshow(img)
```

## Edge detection convolutions

```python
fig = plt.figure(figsize=(15,45))
fig.add_subplot(1,3,1)
plt.imshow(roberts(img))
fig.add_subplot(1,3,2)
plt.imshow(prewitt(img))
fig.add_subplot(1,3,3)
plt.imshow(sobel(img))
```

## Gaussian filter

```python
fig = plt.figure(figsize=(15, 45))
fig.add_subplot(1,3,1)
plt.imshow(img)
fig.add_subplot(1,3,2)
plt.imshow(gaussian(img, kernel_size=3))
fig.add_subplot(1,3,3)
plt.imshow(gaussian(img, kernel_size=7, sigma=2))
```

## Mean filter

```python
fig = plt.figure(figsize=(15, 45))
fig.add_subplot(1,3,1)
plt.imshow(img)
fig.add_subplot(1,3,2)
plt.imshow(mean(img, kernel_size=3))
fig.add_subplot(1,3,3)
plt.imshow(mean(img, kernel_size=7))
```

## Median filter

```python
fig = plt.figure(figsize=(15, 45))
fig.add_subplot(1,3,1)
plt.imshow(img)
fig.add_subplot(1,3,2)
plt.imshow(median(img, kernel_size=3))
fig.add_subplot(1,3,3)
plt.imshow(median(img, kernel_size=7))
```
